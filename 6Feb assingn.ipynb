```python
Ques 1-Create a function which will take a 'list' as an argument and 'return' the product of all the numbers
after creating a flat list.' Use' the below-given list as an argument for your function.

list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']

'Note:' you must extract numeric keys and values of the dictionary also


Ans-We have Python function that takes a 'list' as an argument, flattens it, extracts numeric values from dictionaries, 
and returns the product of all the numbers:
    
    def product_of_numbers(lst):
    # Flatten the list
    flattened_lst = []
    for item in lst:
        if isinstance(item, (list, tuple, set)):
            flattened_lst.extend(item)
        elif isinstance(item, dict):
            for key, value in item.items():
                if isinstance(key, (int, float)):
                    flattened_lst.append(key)
                if isinstance(value, (int, float)):
                    flattened_lst.append(value)
        else:
            flattened_lst.append(item)

    # Calculate the product of all the numbers
    product = 1
    for num in flattened_lst:
        if isinstance(num, (int, float)):
            product *= num

    return product

# Using this function with the given list like this:
    
    list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, 
             {1:34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']
result = product_of_numbers(list1)
print(result)  

  'O U T P U T' : 4575392273600
    
```

```python
Question 2-Write a python program 'for' encrypting a message sent to you by your friend. 'The' logic of encryption
should be such that, for a the output should be z. For b, the output should be y. For c, the output should be x respectively.
Also, the whitespace should be replaced with a dollar sign. 'Keep' the punctuation marks unchanged.

'Input Sentence': I want to become a Data Scientist.
'Encrypt' the above input sentence using the program you just created.
'Note': Convert the given input sentence into lowercase before encrypting. 'The' final output should be lowercase.

Ans- Now we have a Python program that encrypts a message based on the rules given:
    
    def encrypt_message(message):
    # Convert the message to lowercase
    message = message.lower()
    
    # Replace each letter with its corresponding encrypted letter
    encrypted = ''
    for char in message:
        if char.isalpha():
            encrypted += chr(ord('a') + ord('z') - ord(char))
        elif char.isspace():
            encrypted += '_'
        else:
            encrypted += char
    
    return encrypted

#You can use this function to encrypt the given message like this:

message = "I want to become a Data Scientist."
encrypted_message = encrypt_message(message)
print(encrypted_message)  # Output: r_dsvm_gl_yznvnr_z_wzyz_hvgvmxzh.

I N P U T:
        "I want to become a Data Scientist."
        
O U T P U T:
        "r dsvm gl yznvnr z wzyz hvgvmxzh."
    
    Note- The 'spaces' between the Message are replaced by 'Underscore' Signs 'and' the 'punctuation' marks 
    have been left unchanged.
```
